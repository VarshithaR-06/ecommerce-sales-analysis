```python
# Import the tools we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('Online Sales Data.csv')

# Show the first 5 rows
print("Shape of data:", df.shape)
df.head()
```

```python
# Check data types of each column
print("=== Data Types ===")
print(df.dtypes)

# Check for missing values
print("\n=== Missing Values ===")
print(df.isnull().sum())

# Check for duplicate rows
print("\n=== Duplicate Rows ===")
print("Duplicates found:", df.duplicated().sum())
```

```python
# Convert Date from text to proper date format
df['Date'] = pd.to_datetime(df['Date'])

# Extract Month and Year as separate columns
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Month_Name'] = df['Date'].dt.strftime('%B')

# Confirm it worked
print("Date column type is now:", df['Date'].dtype)
print("New columns added:", ['Month', 'Year', 'Month_Name'])
df[['Date','Month','Month_Name','Year']].head()
```

```python
# Verify Total Revenue = Units Sold x Unit Price
df['Calculated_Revenue'] = df['Units Sold'] * df['Unit Price']
df['Revenue_Check'] = round(df['Calculated_Revenue'] - df['Total Revenue'], 2)

print("Revenue mismatches:", (df['Revenue_Check'] != 0).sum())

# Drop the check columns — we don't need them anymore
df.drop(columns=['Calculated_Revenue', 'Revenue_Check'], inplace=True)

print("Data is verified and clean!")
print("Final shape:", df.shape)
print("Final columns:", list(df.columns))
```

```python
# Save cleaned data as a new CSV file
df.to_csv('Online_Sales_Clean.csv', index=False)

print("Clean file saved as: Online_Sales_Clean.csv")
print("Total rows:", len(df))
print("Total columns:", len(df.columns))
print("\nAll columns in clean file:")
for col in df.columns:
    print(" -", col)
```
